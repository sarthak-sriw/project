```python
import numpy as np
import time
import cv2



cap=cv2.VideoCapture(0)

time.sleep(3)

background=0

#capture the best background
for i in range(30):
    ret,background=cap.read()
    
    
while(cap.isOpened()):
    
    ret,img=cap.read()
    
    if not ret:
        break
    
    
    #RGB color space is sensitive to illumination such in case of shadow it 
    #appears brownish or some other color so we cannot generate all color space to detect cloth
    #hue is used to select color
    #saturation select intensity/different shades
    # value --brightness
    hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
    
    #detect cloth
    #red is from 0-30 and 170-180 --to 10 because it do not detect skin color
    lower_red=np.array([0,120,70])
    upper_red=np.array([10,255,255])
    mask1=cv2.inRange(hsv,lower_red,upper_red)
    
    lower_red=np.array([170,120,70])
    upper_red=np.array([180,255,255])
    mask2=cv2.inRange(hsv,lower_red,upper_red)
    
    mask1=mask1+mask2
    
    #remove noises and dilate them
    mask1=cv2.morphologyEx(mask1,cv2.MORPH_OPEN,np.ones((3,3),np.uint8),iterations=2)
    mask1=cv2.morphologyEx(mask1,cv2.MORPH_DILATE,np.ones((3,3),np.uint8),iterations=1)
    
    
    #everything other than cloak
    mask2=cv2.bitwise_not(mask1)
    
    
    #mask decides how to perform operation
    res1=cv2.bitwise_and(background,background,mask=mask1)
    res2=cv2.bitwise_and(img,img,mask=mask2)
    
    final_output=cv2.addWeighted(res1,1,res2,1,0)
    
    cv2.imshow('invisible cloak',final_output)
    
    k = cv2.waitKey(10)
    if k == 27:
        break

        
        
        
cap.release()
cv2.destroyAllWindows()
    
    
    
    
    
    
    
    
    
    
    
    
    
```
